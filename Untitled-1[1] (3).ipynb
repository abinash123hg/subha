# Exploratory The Data Analysis

***(1) Demographic Distribution of Customers.***

```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['0-17', '18-25', '26-35', '36-44', '45-50', '55+']
male = [100, 250, 300, 35, 270, 45]
female = [80, 100, 240, 200, 25, 28]
x = np.arange(len(categories))
width = 0.4 
plt.bar(x, male, width, label='Male', color='blue')
plt.bar(x, female, width, bottom=male, label='Female', color='green')
plt.xlabel('Age Groups',fontsize = 15)
plt.ylabel('Number of Customers',fontsize= 15)
plt.title('Customer Demographics',fontsize = 20)
plt.xticks(x, categories)
plt.legend()
for i in range(len(categories)):
    plt.text(i, male[i] + 10, str(male[i]), ha='center', va='bottom', fontsize=10, color='blue') 
    plt.text(i, male[i] + female[i] + 10, str(female[i]), ha='center', va='bottom', fontsize=10, color='green')
plt.tight_layout()
plt.show()
```

*female customers in the 18-25, 36-44, and 55+ age groups can effectively balance gender distribution of sales.*

***(2) Daily Sales Performance.***

```python
import matplotlib.pyplot as plt
days = ["Monday", "Tuesday", "Wednesday", "{Thursday}", "Friday", "Saturday", "Sunday"]
sales = [3000,8500,15000,24000,13000, 9000,4500]
plt.figure(figsize=(8, 6))
plt.bar(days, sales)
plt.title("Total Sales Per Day", fontsize=16)
plt.xlabel("Days", fontsize=14)
plt.ylabel("Total Sales in price", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=1,color = "black")
plt.show()
```

* from the above graph we can see that highest sales occurring on Thursday and Wednesday.*

***(3) Sales Distribution by Product Category***


```python
import matplotlib.pyplot as plt
categories = ["Lakshmi Bomb","Rocket","Phuljhari","Sky Shot","crackles"]
sales = [25, 30, 20, 15, 10] 
plt.figure(figsize=(10, 10))
plt.pie(
    sales,
    labels=categories,
    autopct="%1.1f%%",
    colors=["yellow", "lightcoral", "hotpink", "green", "red"],
)
plt.title("Product Category Sales Distribution", fontsize=23)
plt.show()
```

*From the above graph we can see that Rocket and Lakshmi Bomb are the top-selling products,and VIP Bomb are least-selling product.*

***(4) Exploring Diwali Sales by State***

```python
import matplotlib.pyplot as plt
states = ['Maharashtra', 'Karnataka', 'Delhi', 'Gujarat', 'Tamil Nadu', 'West Bengal']
sales = [5000000,20000000,10000000,50000000,30000000,60000000] 
plt.figure(figsize=(10, 6))
plt.bar(states, sales, color='orange', edgecolor='black')
plt.title('State-wise Diwali Sales', fontsize=20)
plt.xlabel('States', fontsize=15)
plt.ylabel('Sales in CR', fontsize=15)
plt.yticks(ticks=[5000000,20000000,10000000,50000000,30000000,60000000],
labels=["Five Crore","Twenty Crore","Ten Crore","Fifty Crore","Thirty Crore","Sixty Crore"])
plt.xticks(rotation=35, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=1,color = "black")
plt.tight_layout()
plt.show()
```

*From above graphs we can see that most of the total sales/amount are from West Bengle,Tamil Nadu and Gujarat respectively*

***(5) Analysis of Sales Revenue.***

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Date': ['2024-10-24', '2024-10-25', '2024-10-26', '2024-10-27', '2024-10-28', '2024-10-29','2024-10-30'],
    'Sales Revenue': [2000,5000,10000,15000,19000, 24000,19000]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales Revenue'], marker='o', linestyle='--', color='red')
plt.title('Sales Revenue Over Time', fontsize=20)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Sales Revenue', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.6,color = "black")
plt.show()
```

*From the above graph we can see that the revenue increases from 2,000 INR on October 24 to 24,000 INR on October 30.*

***(6) Monthly Financial Performance.***

```python
import matplotlib.pyplot as plt
categories = ['september', 'october', 'november', 'december']
profits = [8000,15000,5000,-6000]
plt.figure(figsize=(12, 6)) 
plt.fill_between(categories, profits, color="blue", alpha=1)
plt.xlabel('Months',fontsize = 15)
plt.ylabel('Loss   |   Profit',fontsize = 15)
plt.title('Monthly Profit and Loss ',fontsize = 20)
plt.grid(True, linestyle='--', alpha=0.6,color = "black")
plt.tight_layout()
plt.show()
```

*from the above graph we can see october experienced the highest profit,november and december saw losses,*
